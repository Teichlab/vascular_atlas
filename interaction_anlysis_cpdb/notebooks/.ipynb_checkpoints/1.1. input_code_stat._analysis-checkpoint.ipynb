```python
adata.obs['org_celltype'] = adata.obs['organ_uni'].astype(str) + "_" + adata.obs['ann_final_2'].astype(str)
```

```python
adata.obs['org_celltype'].value_counts()
```

```python
adata.obs['celltype'] = adata.obs['org_celltype']
# subset only to categories with more than 15 cells
adata = adata[adata.obs['celltype'].map(adata.obs['celltype'].value_counts()) > 20]
adata
```

```python
adata.obs['Organ'] = adata.obs['organ_uni']
adata.obs['Organ']
adata.obs['Organ'] = adata.obs['Organ'].str.capitalize()
adata.obs['Organ_celltype'] = np.asarray(adata.obs['Organ']) + '_' +  np.asarray(adata.obs['ann_final_2'])
```

```python
#minimise metadata
adata.obs = adata.obs[['organ_uni','donor','kit', 'sorting','location','age','gender', 'dissociation',
                       'ann_ec_states', 'ann_mural_final', 'ann_final', 'ann_final_2', 'Organ', 'Organ_celltype']]
adata.obs['celltype'] = adata.obs['Organ_celltype']
adata.obs_names_make_unique()
adata
```
