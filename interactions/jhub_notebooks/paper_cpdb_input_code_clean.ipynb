# Prepare files for cpdb

```python
import scanpy as sc
adata = sc.read_h5ad('/home/jovyan/1_FIBROSIS/adata_for_cpdb.h5ad')
```

```python
import pandas as pd
# export the counts for cpdb
adata_counts = pd.DataFrame(adata.X.todense(), index=adata.obs_names, 
                                 columns=adata.var_names)
#tranpose
adata_counts = adata_counts.T
adata_counts
```

```python
adata_counts.to_csv('/home/jovyan/1_FIBROSIS/cpdb_counts.csv')
```

```python
#rename "celltype" category with the column of interest for cpdb run
adata_meta = adata.obs
adata_meta['celltype'] = adata_meta['Organ_celltype']
# export metadata for cpdb - too much metadata here, can do with less
adata_meta.to_csv('/home/jovyan/1_FIBROSIS/cpdb_meta.csv')
```

```python
#to make the barcodes in meta consistent, for some reason the barcodes change from '.' to '-'
#rownames(meta) = gsub('[-]', '.', rownames(meta))
```

```python
import pandas as pd
df_meta = pd.DataFrame(data={'Cell':list(adata.obs.index),
                             'celltype':[ i for i in adata.obs['celltype']],
                             'Organ':[ i for i in adata.obs['Organ']],          
                             'Organ_celltype':[ i for i in adata.obs['Organ_celltype']], 
                            })
df_meta.set_index('Cell', inplace=True)
df_meta.to_csv('/home/jovyan/1_FIBROSIS/meta.tsv', sep = '\t')
```

```python
df_meta = pd.DataFrame(data={'celltype':[ i for i in adata.obs['celltype']],
                             'microenvironment':[ i for i in adata.obs['Organ']]})
df_meta.to_csv('/home/jovyan/1_FIBROSIS/microenvironments.tsv', sep = '\t',index=False)
```
